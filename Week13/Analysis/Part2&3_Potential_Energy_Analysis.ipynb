# Make a plot of potential energy vs time in the NVT and NPT energy files in Outputs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data in the NVT energy files
Step, Time, Kin,Temp,Pot,ConsQty,UsedTime = np.loadtxt('../Outputs/NVT-1.ener', skiprows=1, unpack=True)
```

```python
# plot the potential energy in the NVT energy files
plt.plot(Time, Pot)
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u.)')
plt.title('Potential Energy vs. Time in NVT')
plt.savefig(f'../Figures/Potential_Energy_vs_Time_NVT.pdf',format="pdf")
print('Plot saved: ../Figures/Potential_Energy_vs_Time_NVT.pdf')

plt.show()
```

```python
# load data in the NPT energy files
Step, Time, Kin,Temp,Pot,ConsQty,UsedTime = np.loadtxt('../Outputs/NPT-1.ener', skiprows=1, unpack=True)
```

```python
# plot the potential energy in the NPT energy files
plt.plot(Time, Pot)
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u.)')
plt.title('Potential Energy vs. Time in NVT')
plt.savefig(f'../Figures/Potential_Energy_vs_Time_NPT.pdf',format="pdf")
print('Plot saved: ../Figures/Potential_Energy_vs_Time_NPT.pdf')

plt.show()
```
