Plot how each replica is moving in temperature in replica_temp 
(you may have to do every 100 frame to see what's happening) and make sure it goes up and down

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load the raw data and check
time, index300, index400, index500, index600  = np.loadtxt('../Inputs/ParallelTemp2nd/T300/replica_temp.xvg', unpack=True)
print(time.shape)
print(index300)
print(index400)
print(index500)
print(index600.shape)

# Extract one from every 100 frame
def frame_extraction(index):
    n=index.size
    #print(n)
    index = index[np.arange(0,n,100)]
    #print(index_dele)
    #print(index.shape)
    return index

time = frame_extraction(time)
index300 = frame_extraction(index300)
index400 = frame_extraction(index400)
index500 = frame_extraction(index500)
index600 = frame_extraction(index600)

# recheck
print("-----------------extarction checking-----------------")
print(time.shape)
print(index300)
print(index400)
print(index500)
print(index600.shape)
  
```

```python
# Plot the Replica Exchange
fig,(ax1,ax2,ax3,ax4) =plt.subplots(4,1, figsize=(7,9))
ax1.plot(time,index300,label='Replica1: T=300K',color='blue')
ax2.plot(time,index400,label='Replica2: T=400K',color='green')
ax3.plot(time,index500,label='Replica3: T=500K',color='red')
ax4.plot(time,index600,label='Replica4: T=600K',color='black')
# adjust 
ax1.legend(loc=(1.05, 0.5))
ax2.legend(loc=(1.05, 0.5))
ax3.legend(loc=(1.05, 0.5))
ax4.legend(loc=(1.05,0.5))
ax4.set_xlabel("Time(ps)", fontsize=15)


# set titles
plt.text(x= -900, y=8, s="Temperature(K)", fontsize=15, va="center", rotation=90)
plt.text(x=0.5, y=0.85, s="Replica Exchange", fontsize=20, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s= "Parallel Tempering of Alanine Dipeptide(ADP)", fontsize=13, ha="center", transform=fig.transFigure)
# adjust 
plt.subplots_adjust(top=0.8, wspace=0.3)
# save figure
plt.savefig("../Figures/ADP_ReplicaExchange_ParallelTempering.pdf",format="pdf")
print("Plot saved: ../Figures/ADP_ReplicaExchange_ParallelTempering.pdf")
plt.show()

```
