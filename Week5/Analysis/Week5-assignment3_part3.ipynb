Compute a histogram of energy in each case, and plot all the histograms on the same plot. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

time300,T300=np.loadtxt('../Inputs/ParallelTemp2nd/T300/energy.xvg', skiprows=24, unpack=True)
time400,T400=np.loadtxt('../Inputs/ParallelTemp2nd/T400/energy.xvg',skiprows=24, unpack=True)
time500,T500=np.loadtxt('../Inputs/ParallelTemp2nd/T500/energy.xvg',skiprows=24, unpack=True)
time600,T600=np.loadtxt('../Inputs/ParallelTemp2nd/T600/energy.xvg',skiprows=24, unpack=True)

print(time300)
print(time300.shape)
print(T300)
print(T300.shape)
```

```python
# plot potential energy vs time
fig = plt.figure(5)
# plot replica 1, colorblind-friendly, label
plt.hist(T300, bins=100, color='g', label='replica1: T = 300 K', alpha=1, density=True)
# plot replica 2, colorblind-friendly, label
plt.hist(T400, bins=100, color='r', label='replica2: T = 400 K', alpha=0.8, density=True)
# plot replica 3, colorblind-friendly, label
plt.hist(T500, bins=100, color='b', label='replica3: T = 500 K', alpha=0.6, density=True)
# plot replica 4, colorblind-friendly, label
plt.hist(T600, bins=100, color='y', label='replica4: T = 600 K', alpha=0.4, density=True)
# legend, fontsize = 11, relocate to the right
plt.legend(loc='upper right', fontsize=11, frameon=False)

# set titles
#plt.text(x= 1, y=8, s="Probability", fontsize=15, va="center", rotation=90)
plt.text(x=0.5, y=0.95, s="Potential Energy Distribution", fontsize=20, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.9, s= "Parallel Tempering of Alanine Dipeptide(ADP)", fontsize=13, ha="center", transform=fig.transFigure)
plt.xlabel('Potential energy(kJ/mol)', fontsize=12)
plt.ylabel('Probability', fontsize=12)
# save figure
plt.savefig("../Figures/ADP_ReplicaExchange_ParallelTempering.pdf",format="pdf")
print("Plot saved: ../Figures/ADP_ReplicaExchange_ParallelTempering.pdf")
plt.show()

```
