# rmsd of each predicted strucutre based on alpha carbon atoms, backbone and sidechain
Using mdtraj, in a jupyter notebook, align your alphafold predictions to the crystal structure 
and make a bar plot showing the RMSD of each predicted structure to the crystal structure based on alpha-carbons and based on the full backbone and side-chains. 
Does alphafold get everything right or just the backbone structure?

```python
# load libraries
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load alphafold predictions
alphafold_1= md.load_pdb('../data/2W9T/unrelaxed_model_1_pred_0.pdb')
alphafold_2= md.load_pdb('../data/2W9T/unrelaxed_model_2_pred_0.pdb')
alphafold_3= md.load_pdb('../data/2W9T/unrelaxed_model_3_pred_0.pdb')
alphafold_4= md.load_pdb('../data/2W9T/unrelaxed_model_4_pred_0.pdb')
alphafold_5= md.load_pdb('../data/2W9T/unrelaxed_model_5_pred_0.pdb')
# load colabfold predictions
colabfold_1 = md.load_pdb('../data/test_596c5_result/test_596c5_unrelaxed_rank_1_model_5.pdb')
colabfold_2 = md.load_pdb('../data/test_596c5_result/test_596c5_unrelaxed_rank_2_model_3.pdb')
colabfold_3 = md.load_pdb('../data/test_596c5_result/test_596c5_unrelaxed_rank_3_model_4.pdb')
colabfold_4 = md.load_pdb('../data/test_596c5_result/test_596c5_unrelaxed_rank_4_model_1.pdb')
colabfold_5 = md.load_pdb('../data/test_596c5_result/test_596c5_unrelaxed_rank_5_model_2.pdb')
print(colabfold_1)

crystal_structure = md.load_pdb('../data/2W9T.pdb')
print(crystal_structure)
```

```python
# load crystal structure, only chain A

crystal_chainA = crystal_structure.atom_slice(crystal_structure.top.select('chainid 0'))

#print()
```

```python
# select alpha carbon atoms
alpha_carbon = 'name CA'
print(alpha_carbon)
# select the full backbone
"backbone" = 'name CA'
# select the sidechain
sidechain = 'sidechain'
```

```python
# calculate rmsd of each predicted strucutre based on alpha carbon atoms
pdblist=[alphafold_1,alphafold_2,alphafold_3,alphafold_4,alphafold_5,
         colabfold_1,colabfold_2,colabfold_3,colabfold_4,colabfold_5]
rmsd_CA=[]
for i in pdblist:
    rmsd_CA.append(md.rmsd(i, crystal_chainA, atom_indices=crystal_chainA.top.select('name CA')))
print(rmsd_CA)
#A1=md.rmsd(alphafold_5, crystal_chainA, atom_indices=crystal_chainA.top.select('name CA'))
#print(A1)
```

```python
# make a bar plot for the rmsd of each predicted strucutre based on alpha carbon atoms
fig=plt.figure(figsize=(15, 5))

x_labels = ['alphafold_1', 'alphafold_2', 'alphafold_3', 'alphafold_4', 'alphafold_5', 'colabfold_1', 'colabfold_2', 'colabfold_3', 'colabfold_4', 'colabfold_5']
y_values = np.array(rmsd_CA).flatten()
plt.bar(x_labels, y_values)
# set labels
#plt.xlabel('ML predictions', fontsize=12)
plt.ylabel('RMSD (Å)', fontsize=12)
# set all titles
plt.text(x=0.5, y=0.95, s="RMSD of Each Prediction Based on Alpha Carbon Atoms", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.9, s= "Machine learning prediction strucutre of 2W9T generated from Alphafold and Colabfold", fontsize=11, ha="center", transform=fig.transFigure)

# save figure
plt.savefig("../Figures/RMSD_Based_on_Alpha_Carbon_Atoms.pdf",format="pdf")
print("Plot saved: ../Figures/RMSD_Based_on_Alpha_Carbon_Atoms.pdf")
plt.show()
```

```python
# calculate rmsd of each predicted strucutre
pdblist=[alphafold_1,alphafold_2,alphafold_3,alphafold_4,alphafold_5,
         colabfold_1,colabfold_2,colabfold_3,colabfold_4,colabfold_5]
rmsd_backbone=[]
for i in pdblist:
    rmsd_backbone.append(md.rmsd(i, crystal_chainA, atom_indices=crystal_chainA.top.select('backbone')))
print(rmsd_backbone)
```

```python
# make a bar plot for the rmsd of each predicted strucutre based on backbone
fig=plt.figure(figsize=(15, 5))

x_labels = ['alphafold_1', 'alphafold_2', 'alphafold_3', 'alphafold_4', 'alphafold_5', 'colabfold_1', 'colabfold_2', 'colabfold_3', 'colabfold_4', 'colabfold_5']
y_values = np.array(rmsd_backbone).flatten()
plt.bar(x_labels, y_values)
# set labels
plt.ylabel('RMSD (Å)', fontsize=12)
# set all titles
plt.text(x=0.5, y=0.95, s="RMSD of Each Prediction Based on Backbone", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.9, s= "Machine learning prediction strucutre of 2W9T generated from Alphafold and Colabfold", fontsize=11, ha="center", transform=fig.transFigure)

# save figure
plt.savefig("../Figures/RMSD_Based_on_backbone.pdf",format="pdf")
print("Plot saved: ../Figures/RMSD_Based_on_backbone.pdf")
plt.show()
```

```python
# calculate rmsd of each predicted strucutre based on sidechain
pdblist=[alphafold_1,alphafold_2,alphafold_3,alphafold_4,alphafold_5,
         colabfold_1,colabfold_2,colabfold_3,colabfold_4,colabfold_5]
rmsd_sidechain=[]
for i in pdblist:
    rmsd_sidechain.append(md.rmsd(i, crystal_chainA, atom_indices=crystal_chainA.top.select('sidechain')))
print(rmsd_sidechain)

```

```python
# make a bar plot for the rmsd of each predicted strucutre based on backbone
fig=plt.figure(figsize=(15, 5))

x_labels = ['alphafold_1', 'alphafold_2', 'alphafold_3', 'alphafold_4', 'alphafold_5', 'colabfold_1', 'colabfold_2', 'colabfold_3', 'colabfold_4', 'colabfold_5']
y_values = np.array(rmsd_sidechain).flatten()
plt.bar(x_labels, y_values)
# set labels
plt.ylabel('RMSD (Å)', fontsize=12)
# set all titles
plt.text(x=0.5, y=0.95, s="RMSD of Each Prediction Based on sidechain", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.9, s= "Machine learning prediction strucutre of 2W9T generated from Alphafold and Colabfold", fontsize=11, ha="center", transform=fig.transFigure)

# save figure
plt.savefig("../Figures/RMSD_Based_on_sidechain.pdf",format="pdf")
print("Plot saved: ../Figures/RMSD_Based_on_sidechain.pdf")
plt.show()
```

```python
fig=plt.figure(figsize=(15, 5))
barWidth = 0.25
# make a bar plot
bars1 = np.array(rmsd_CA).flatten()
bars2 = np.array(rmsd_backbone).flatten()
bars3 = np.array(rmsd_sidechain).flatten()
x_labels = ['alphafold_1', 'alphafold_2', 'alphafold_3', 'alphafold_4', 'alphafold_5', 'colabfold_1', 'colabfold_2', 'colabfold_3', 'colabfold_4', 'colabfold_5']
plt.bar(x_labels, bars3, color='#FF5511', width=barWidth, edgecolor='white', label='Based_on_sidechain')
plt.bar(x_labels, bars2, color='#00BBFF', width=barWidth, edgecolor='white', label='Based_on_backbone')
plt.bar(x_labels, bars1, color='#FF0088', width=barWidth, edgecolor='white', label='Based_on_Alpha_Carbon_Atoms')
# set labels
plt.legend()
plt.ylabel('RMSD (Å)', fontsize=12)
# set all titles
plt.text(x=0.5, y=0.95, s="RMSD of Each Prediction", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.9, s= "Machine learning prediction strucutre of 2W9T generated from Alphafold and Colabfold", fontsize=11, ha="center", transform=fig.transFigure)

# save figure
plt.savefig("../Figures/RMSD_Based_on_sidechain_backbone_CA.pdf",format="pdf")
print("Plot saved: ../Figures/RMSD_Based_on_sidechain_backbone_CA.pdf")
plt.show()
```
