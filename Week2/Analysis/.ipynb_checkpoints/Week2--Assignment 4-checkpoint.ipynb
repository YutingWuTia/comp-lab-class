### b) Make 6 separate plots for each quantity vs time, with time in ns

```python

# convert time unit from ps to ns
time_ns = [i * 1e-3 for i in time]
print(len(time_ns))


def quantity_vs_time(quantity_array, quantity_name, quantity_unit):
    
    ## Draw time(x-axis) and quantity(y-axis) in the chart, add label
    plt.plot(time_ns, quantity_array, label= f'{quantity_name}',linewidth=0.5)     
    plt.xlabel('Time(ns)',fontsize=12)                                        ## Add x-axis label
    plt.ylabel(f'{quantity_name}({quantity_unit})',fontsize=12)               ## Add y-axis label
    plt.title(f'{quantity_name} VS Time',fontsize=20)                         ## Set the title of the chart
    plt.show()                                                                ## plot the chart
   
```

```python
quantity_vs_time(pot_e, 'Potential_energy', 'kJ/mol')
quantity_vs_time(kin_e, 'Kinetic_energy', 'kJ/mol')
quantity_vs_time(tot_e, 'Total_energy', 'kJ/mol')
quantity_vs_time(temp, 'Temperature', 'K')
quantity_vs_time(press, 'Pressure', 'bar')
quantity_vs_time(vol, 'Volumn', 'nm^3')
```

### c) Make a plot that has potential energy and temperature on the same plot
##### where the left axis is potential energy and the right axis is kinetic energy, using different colors and different axis labels

```python
#define colors to use
col1 = 'blue'                  
col2 = 'red'                    


Potential_plot,Kinetic_plot = plt.subplots()                                       #define subplots
Kinetic_plot.plot(time_ns, kin_e, color=col1,linewidth=0.5)                        #add first line to plot
Kinetic_plot.set_xlabel('Time(ns)', fontsize=12)                                   #add x-axis label
Kinetic_plot.set_ylabel('Kinetic_energy(kJ/mol)', color=col1, fontsize=12)         #add y-axis label
#Kinetic_plot.show
kin_pot_plot = Kinetic_plot.twinx()                                                #define second y-axis that shares x-axis with current plot
kin_pot_plot.plot(time_ns, pot_e, color=col2,linewidth=0.5)                        #add second line to plot
kin_pot_plot.set_ylabel('Potential_energy(kJ/mol', color=col2, fontsize=12)        #add second y-axis label
```
