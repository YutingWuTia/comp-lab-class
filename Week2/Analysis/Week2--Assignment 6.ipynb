### a) read in the md trajectory of beta amyloid

```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

```python
traj = md.load_xtc('../Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='../Data/1hz3_T310.start.nowater.gro')
traj
```

### b) compute the end-end distance at each time, and the radius of gyration at each time

```python
# compute the distance between the first and last atom of each frame
def end_end(traj):
    # get the first and last atom
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atom
    dist = np.sqrt(np.sum((first-last)**2, axis=1))
    return dist

# compute the distance between the first and last atom
dist = end_end(traj)
print(dist)
# the simulation time (in picoseconds)
print(traj.time)
```

```python
# Compute the radius of gyration for every frame.
Rgndarray = md.compute_rg(traj, masses=None)
Rgndarray
```

### c) plot the end-end distance and radius of gyration vs time on the same plot, clearly labeled (same y axis is fine this time)

```python
plt.plot(traj.time, dist, label= "End to End Distance")           #plot distance vs time
plt.plot(traj.time, Rgndarray, label= "Radius of Gyration")       #plot radius of gyration vs time
plt.legend()                     
plt.xlabel('Time (ps)')
plt.ylabel('(nm)')
plt.title('Time VS End to End Distance and Radius of Gyration')
plt.show()
```

### d) plot a normalized histogram of end-end distance and another of radius of gyration

```python
# normalization of end-end distance
print(dist)
dist_norm = dist/np.max(dist)
print(dist_norm)
# normalization of radius of gyration
print(Rgndarray)
Rgndarray_norm = Rgndarray/np.max(Rgndarray)
print(Rgndarray_norm)
```

```python
# plot the normalized histogram of end-end distance
plt.hist(dist_norm, bins=10, edgecolor='black')                     
plt.xlabel('end-end distance')
plt.ylabel('number')
plt.title('normalized end-end distance')
plt.show()
```

```python
# plot the normalized histogram of radius of gyration
plt.hist(Rgndarray_norm, bins=10, edgecolor='black')                     
plt.xlabel('radius of gyratio')
plt.ylabel('number')
plt.title('normalized radius of gyratio')
plt.show()
```
