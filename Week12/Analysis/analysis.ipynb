# Plot the energy versus time

```python
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.analysis.rdf import InterRDF
import MDAnalysis as mda
```

```python
# for water system

# load data
Step, Time, Kin,Temp,Pot,ConsQty,UsedTime = np.loadtxt('../Runs/Water/water_aimd_1ps_every0.5fs-1.ener', skiprows=1, unpack=True)
```

```python
# plot the potential energy of water
plt.plot(Time, Pot)
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u.)')
plt.title('Potential Energy vs. Time for Water')
plt.savefig(f'../Figures/Potential_Energy_vs_Time_Water.pdf',format="pdf")
print('Plot saved: ../Figures/Potential_Energy_vs_Time_Water.pdf')

plt.show()
```

```python
# plot the kinetic energy of water
plt.plot(Time, Kin)
plt.xlabel('Time (fs)')
plt.ylabel('kinetic Energy (a.u.)')
plt.title('kinetic Energy vs. Time for Water')
plt.savefig(f'../Figures/kinetic_Energy_vs_Time_Water.pdf',format="pdf")
print('Plot saved: ../Figures/kinetic_Energy_vs_Time_Water.pdf')

plt.show()
```

# Use mdtraj or mdanalysis to compute g(r) for the oxygens

```python
# g(r) for Oxygen in water system
# The box size for the water system is: water with unit cell: ABC [angstrom] 12.42 12.42 12.42

# load trajectory
traj = mda.Universe('../Runs/Water/water_aimd_1ps_every0.5fs-pos-1.xyz')
traj

# set dimensions of the box
traj.dimensions = np.array([12.42, 12.42, 12.42, 90, 90, 90])
```

```python
# calculate the radial distribution function for Oxygen
box_size = 12.42
rdf = InterRDF(traj.select_atoms('name O'), traj.select_atoms('name O'), range=(0.0, box_size / 2), nbins=100)
rdf.run()
```

```python
# plot the radial distribution function for Oxygen
plt.plot(rdf.bins[1:], rdf.rdf[1:])

plt.xlabel('Distance (Angstrom)')
plt.ylabel('g(r)')
plt.title('Radial Distribution Function for Oxygen in water')
plt.savefig(f'../Figures/Radial_Distribution_Function_for_Oxygen_water.pdf',format="pdf")
print('Plot saved: ../Figures/Radial_Distribution_Function_for_Oxygen_water.pdf')


plt.show()
# the first one shouldn't be included because it's the RDF of the atom with itself
```
