```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# copy data from Week10/Part2/Part2.ipynb
Rvals = [3.5, 3.75, 4.0, 4.25, 4.5, 4.75, 5.0, 5.25, 5.5, 5.75, 6.0]
Energy = [0.168852, -0.069952, -0.123122, -0.116537, -0.095162, -0.073663, -0.055927, -0.042277, -0.032054, -0.024469, -0.018843]
```

```python
def lj(r,sigma,eps):
    return 4*eps*( (sigma/r)**12-(sigma/r)**6  )
```

```python
# Make a plot of the energy versus distance 
# with the computed values as dots and the values from the fit function with a dashed line

fig = plt.figure()
# scatter plot
plt.scatter(Rvals,Energy,label='CCSD')

# fit data
popt, pcov = curve_fit(lj, Rvals, Energy)
# plot fit
r = np.linspace(3.5,6.0,100)
plt.plot(r,lj(r,*popt),'--',color='r',label='fit function')


plt.legend(fontsize=12)
plt.xlabel('r (Ã…)',fontsize=12)
plt.ylabel('Energy (kcal/mol)',fontsize=12)
plt.title('Ar-Ar Interaction\n psi4')

plt.savefig('Ar_Ar_Interaction.pdf',format="pdf")
print("Plot saved: ../Figures/Ar_Ar_Interaction.pdf")
plt.show()
```
