```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
phi_biasf10, freeE_biasf10, der_phi_biasf10 = np.loadtxt("/home/yw5806/comp-class/comp-lab-class/Week6/Inputs/assignment_3/adp.fes.dat", skiprows=5, unpack=True)

phi_biasf1, freeE_biasf1, der_phi_biasf1 = np.loadtxt("/home/yw5806/comp-class/comp-lab-class/Week6/Inputs/assignment_5(optional)/bias1/adp_stateA_biasfactor1_pace500_height1.2_sigma0.55_1.05_10ns.fes.dat", skiprows=5, unpack=True)
phi_biasf1·5, freeE_biasf1·5, der_phi_biasf1·5 = np.loadtxt("/home/yw5806/comp-class/comp-lab-class/Week6/Inputs/assignment_5(optional)/bias1.5/adp_stateA_biasfactor1.5_pace500_height1.2_sigma0.55_10ns.fes.dat", skiprows=5, unpack=True)
phi_biasf2, freeE_biasf2, der_phi_biasf2 = np.loadtxt("/home/yw5806/comp-class/comp-lab-class/Week6/Inputs/assignment_5(optional)/bias2/adp_stateA_biasfactor2_pace500_height1.2_sigma0.55_10ns.fes.dat", skiprows=5, unpack=True)
phi_biasf5, freeE_biasf5, der_phi_biasf5 = np.loadtxt("/home/yw5806/comp-class/comp-lab-class/Week6/Inputs/assignment_5(optional)/bias5/adp_stateA_biasfactor5_pace500_height1.2_sigma0.55_10ns.fes.dat", skiprows=5, unpack=True)
phi_biasf20, freeE_biasf20, der_phi_biasf20 = np.loadtxt("/home/yw5806/comp-class/comp-lab-class/Week6/Inputs/assignment_5(optional)/bias20/adp_stateA_biasfactor20_pace500_height1.2_sigma0.55_10ns.fes.dat", skiprows=5, unpack=True)

# check data we are going to plot
print('phi_biasf10:', phi_biasf10.shape)
print('freeE_biasf10:', freeE_biasf10.shape)

print('phi_biasf1:', phi_biasf1.shape)
print('freeE_biasf1:', freeE_biasf1.shape)

print('phi_biasf1·5:', phi_biasf1·5.shape)
print('freeE_biasf1·5:', freeE_biasf1·5.shape)

print('phi_biasf2:', phi_biasf2.shape)
print('freeE_biasf2:', freeE_biasf2.shape)

print('phi_biasf5:', phi_biasf5.shape)
print('freeE_biasf5:', freeE_biasf5.shape)

print('phi_biasf20:', phi_biasf20.shape)
print('freeE_biasf20:', freeE_biasf20.shape)
```

```python
#define colors to use
colors = ['r', 'g', 'b', 'violet', 'aqua', 'black']                   
fig=plt.figure(2)

# plot

plt.plot(phi_biasf1, freeE_biasf1, color=colors[1], label='BIASFACTOR=1')
plt.plot(phi_biasf1·5, freeE_biasf1·5, color=colors[2], label='BIASFACTOR=1·5')
plt.plot(phi_biasf2, freeE_biasf2, color=colors[3], label='BIASFACTOR=2')
plt.plot(phi_biasf5, freeE_biasf5, color=colors[4], label='BIASFACTOR=5')
plt.plot(phi_biasf10, freeE_biasf10, color=colors[0], label='BIASFACTOR=10')
plt.plot(phi_biasf20, freeE_biasf20, color=colors[5], label='BIASFACTOR=20')
# legend
plt.legend(fontsize=10)
# set all labels
plt.xlabel(r'$\phi$ (rad)', fontsize=12)
plt.ylabel('Free energy (kJ/mol)', fontsize=12)

# set all titles
plt.text(x=0.6, y=1, s="Comparison of the free energy for different biasfactors", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.95, s= "Calculated every 100 Gaussian kernels deposited", fontsize=11, ha="center", transform=fig.transFigure)


# save figure
plt.savefig("../Figures/ADP_FreeEnergy_diffBiasfactor_comparison.pdf",format="pdf")
print("Plot saved: ../Figures/ADP_FreeEnergy_diffBiasfactor_comparison.pdf")
plt.show()
```
